# iii) Feature Selection

# Features selection criteria:
1)popularity,budget,revenue,runtime,vote_count,vote_average,release_year,budget_adj,revenue_adj can be used as a metric

2)title can be feature engineered to number of words, thus using as a metric, since movies with some amount of title words are preferred more 

3)genres can be splitted based on | , and encoded thus being used as a metric. but due to large number of genre constraint , onehotencoding will have lot of columns, thus this can be used only if we are taking just genres into account, but since this is a general data, we are omitting it.

4)release date, can be feature engineered to being weekend(fri,sat,sun) or not.

5)production_companies can be used as a metric, but due to various different names, it is feeling better to omit them.

6)homepage can be feature engineered to 0 if no website and 1 , if it has a website

```python
df2=df
df2=df2.drop("cast",axis=1)
df2=df2.drop("genres",axis=1)
df2=df2.drop("director",axis=1)
df2=df2.drop("tagline",axis=1)
df2=df2.drop("overview",axis=1)
df2=df2.drop("keywords",axis=1)
df2=df2.drop("id",axis=1)
df2=df2.drop("production_companies",axis=1)
```

```python
hasHomepage=df2.iloc[:,5].values
numberWords=df2.iloc[:,4].values
releaseDay=df2.iloc[:,7].values
```

```python
length=len(hasHomepage)
for i in range(0,length):
    try:
        if(np.isnan(hasHomepage[i])):
            hasHomepage[i]=0
    except:
        hasHomepage[i]=1
```

```python
for i in range(0,length):
    s=numberWords[i].split()
    numberWords[i]=len(s)
```

```python
from datetime import datetime as dt
```

```python
year=df2.iloc[:,10].values
```

```python
for i in range(0,length):
    releaseDay[i]=releaseDay[i][:-2]+str(year[i])
for i in range(0,length):
    s=releaseDay[i].split('/')
    finalstr=""
    if(len(s[0])<2):
        s[0]="0"+s[0]
    if(len(s[1])<2):
        s[1]="0"+s[1]
    finalstr=s[0]+'/'+s[1]+'/'+s[2]
    t=dt.strptime(finalstr,"%m/%d/%Y")
    x=t.weekday()
    if(x<4):
        releaseDay[i]=0
    else:
        releaseDay[i]=1
    
```

# vi) Removing all the zeroes
Imputing WILL not work in case of budget ,revenue,runtime and release_year. so we have to drop rows which has those as zeroes.

```python
j=0
length=len(df2)
while(j<length):
    row=df2.iloc[j:j+1,:].values
    if(row[0][1]==0 or row[0][2]==0 or row[0][5]==0 or row[0][9]==0 or row[0][10]==0 or row[0][11]==0 ):
        df2.drop(df2.index[j:j+1],inplace=True)
        length-=1
    else:
        j+=1
```

# Movies with the third-lowest and third-highest budget

```python
largest=df2.nlargest(3,["budget"])
largest_inf=df2.nlargest(3,["budget_adj"])
smallest=df2.nsmallest(3,["budget"])
smallest_inf=df2.nsmallest(3,["budget_adj"])
```

```python
def indToTitle(index,df):
    length=len(df)
    for i in range(0,length):
        if(df.iloc[i,1]==index):
            return df.iloc[i][5]
```

```python
print("Title of 3rd largest budget movie: ",indToTitle(largest.index[2],df))
print("Title of 3rd largest budget movie taking inflation into account: ",indToTitle(largest_inf.index[2],df))
print("Title of 3rd smallest budget movie: ",indToTitle(smallest.index[2],df))
print("Title of 3rd smallest budget movie taking inflation into account: ",indToTitle(smallest_inf.index[2],df))
```

# Movies with the most and least earned revenue?

```python
largest=df2.nlargest(1,["revenue"])
largest_inf=df2.nlargest(1,["revenue_adj"])
smallest=df2.nsmallest(1,["revenue"])
smallest_inf=df2.nsmallest(1,["revenue_adj"])
```

```python
print("Title of largest revenue movie: ",indToTitle(largest.index[0],df))
print("Title of largest revenue movie taking inflation into account: ",indToTitle(largest_inf.index[0],df))
print("Title of smallest revenue movie: ",indToTitle(smallest.index[0],df))
print("Title of smallest revenue movie taking inflation into account: ",indToTitle(smallest_inf.index[0],df))
```

# What is the average runtime of movies in the year 2006?


#### We are referring to initial database as many rows were removed in preprocessed dataset which had viable runtime.

```python
length=len(df)
rt=df.iloc[:,12].values
yr=df.iloc[:,-3].values
s=0
c=0
for i in range(0,length):
    if(yr[i]==2006 and rt[i]!=0 ):
        s+=rt[i]
        c+=1
print("Average runtime of movies that came out in year 2006: ",int(s/c)," Minutes")
```

# Average number of words in movie titles between the years 2000-2005

```python
length=len(df)
title=df.iloc[:,5].values
yr=df.iloc[:,-3].values
s=0
c=0
for i in range(0,length):
    if(yr[i]>=2001 and yr[i]<2006 ):
        s+=len(title[i].split())
        c+=1
print("Average number of words in movie titles for movies that came out between 2001-2005: ",int(s/c)," words")
```
