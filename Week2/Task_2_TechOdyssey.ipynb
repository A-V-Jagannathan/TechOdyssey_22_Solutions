```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("imdb-movies.csv")
df2=pd.read_csv("Preprocessed_imdb-movies.csv")
df3=pd.read_csv("NonScaled_final.csv")
```

### 1)Find the average runtime of all movies

```python
x=df.iloc[:,-9].values
su=0
count=0
for i in x:
    if(i!=0):
        su+=i
        count+=1
print("Average runtime of all movies is : ",su/count," minutes")
```

### 2)Plot a boxplot between the release_year and revenue_adj columns.
##### Note: for this question im using the preprocessed dataset from task 1.

```python
sns.set_style("whitegrid")
sns.boxplot(x="release_year",y="revenue_adj",data=df3)
plt.title("Release year vs Revenue")
```

### 3)Construct a scatter plot for the column's revenue and vote count.

```python
x=df3.iloc[:,3].values
y=df3.iloc[:,-5].values
plt.scatter(x,y)
plt.xlabel("Revenue")
plt.ylabel("Vote Count")
plt.title("Revenue vs VoteCount")
plt.show()
```

### 4)Plot a line graph depicting the number of movies released every year.

```python
yrs=list(df.iloc[:,-3].values)
###List of release years of movies
x=[]
y=[]
for i in yrs:
    if i not in x:
        x.append(i)
x=sorted(x)
###List of all years present in dataset
y=[yrs.count(t) for t in x]
###Number of times each year appear in dataset
plt.plot(x,y)
plt.xlabel("Year")
plt.ylabel("Number of movies")
plt.title("Movies per year-line plot")
```

### 5)Find a top director(the one who has made a lot of revenue)

```python
directors=df.iloc[:,8].values
revenue=df.iloc[:,4].values
d={}
for i in range(0,len(directors)):
    if(directors[i] not in d.keys()):
        d[directors[i]]=revenue[i]
    else:
        d[directors[i]]+=revenue[i]
d=sorted(d.items(),key= lambda x: x[1],reverse=True)
print("The Top 3 directors and the revenue they made are \n",d[0],"\n",d[1],"\n",d[2])
```

### 6)Plot a bar plot to find the most-rated movie
#### Note: We are using the original dataset and not the preprocessed one

```python
df.set_index("original_title",inplace=True)
x=df.nlargest(10,["vote_average"])
x=x.iloc[:,[-4]]
x.plot.bar()
```
